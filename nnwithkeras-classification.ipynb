```python
# Create simulated dataset for classification
```

```python
import numpy
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples = 200, n_features = 10, n_classes = 2)
```

```python
print ("Number of examples in the Dataset = ", X.shape[0])
print ("Number of features for each sample = ", X.shape[1])
print ("Possible output classes = ", numpy.unique(y))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state=10)
```

```python
from keras.models import Sequential
model = Sequential()
```

```python
from keras.layers import Dense, Activation
model.add(Dense(10, activation='tanh', input_dim = 10))
```

```python
model.add(Dense(5, activation = 'tanh'))
```

```python
model.add(Dense(1, activation = 'sigmoid'))
```

```python
model.compile(optimizer = 'sgd', loss= 'binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs = 100, batch_size = 5, verbose = 1)
```

```python
y_predicted = model.predict(X[0:10,:])
```

```python
print("Predicted probability for each of the examples belonging to class1: "),
print(y_predicted)
print("Predicted class label for each of the examples: "),
print(numpy.round(y_predicted))
```

```python
model.evaluate(X_train,y_train,batch_size=None,verbose = 1)
```

```python
model.evaluate(X_test,y_test,batch_size=None,verbose = 1)
```
